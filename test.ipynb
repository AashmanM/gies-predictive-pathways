```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("mock_gbm.csv")
df

df.drop(["MAJOR3", "MAJOR4", "MINOR3", "MINOR4"], axis=1, inplace=True)
df.drop(df.index[df["COLL1"] == "BUS"], inplace=True)
df["COLL2"] = df["COLL2"].fillna("None")
df["LastTerm_BM"] = df["LastTerm_BM"].fillna(0)
df["FirstTerm_BM"] = df["FirstTerm_BM"].fillna(0)
df["MAJOR2"] = df["MAJOR2"].fillna("None")
df["MINOR1"] = df["MINOR1"].fillna("None")
df["MINOR2"] = df["MINOR2"].fillna("None")
df["CoreIND"] = df["CoreIND"].replace({"No": "Not started", "Yes": "Started"})
df
```

```python
plt.figure(figsize=(15,10))
major_counts = df["MAJOR1"].value_counts()
major_counts.plot.bar()
plt.show()
```

```python
plt.figure(figsize=(15,10))
minor1_counts = df["MINOR1"].value_counts()
minor1_counts.plot.bar()
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x="CLASS_LEVEL", hue="CoreIND", data=df)
plt.show()
```

```python
plt.figure(figsize=(15,10))
df['CoreIND'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```
